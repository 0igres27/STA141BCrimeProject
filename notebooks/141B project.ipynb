```python
# Odin Nielsen Jupyter Notebook

import pandas as pd
dat = pd.read_csv("cleaned_data_feature_engineered_v1.csv") # read in cleaned dataset
#209, 1569, 5572, 6375

dat = dat.drop(index = 1569) # drop indexes that were assigned as crimes in Davis, but actually were outside of davis
dat = dat.drop(index = 5572)
dat = dat.drop(index = 209)
dat = dat.drop(index = 6375)
```

```python
import folium
import folium.plugins
Assault = folium.FeatureGroup(name = "Assault") # create 27 different groups for the 27 different crimes
Larceny_Theft = folium.FeatureGroup(name = "Larceny/Theft")
All_Other_Offenses = folium.FeatureGroup(name = "All Other Offenses")
Fraud_Offenses = folium.FeatureGroup(name = "Fraud Offenses")
Sex_Offenses_Forcible = folium.FeatureGroup(name = "Sex Offenses-Forcible")
Burglary = folium.FeatureGroup(name = "Burglary")
Weapon_Law_Violations = folium.FeatureGroup(name = "Weapon Law Violations")
Vandalism = folium.FeatureGroup(name = "Vandalism")
Drug_Narcotic_Offenses = folium.FeatureGroup(name = "Drug/Narcotic Offenses")
Driving_Under_the_Influence = folium.FeatureGroup(name = "Driving Under the Influence")
Motor_Vehicle_Theft = folium.FeatureGroup(name = "Motor Vehicle Theft")
Pornography_Obscene_Material = folium.FeatureGroup(name = "Pornography/Obscene Material")
Disorderly_Conduct = folium.FeatureGroup(name = "Disorderly Conduct")
Kidnapping_Abduction = folium.FeatureGroup(name = "Kidnapping/Abduction")
Curfew_Loitering_Vagrancy = folium.FeatureGroup(name = "Curfew/Loitering/Vagrancy")
Stolen_Property = folium.FeatureGroup(name = "Stolen Property")
Trespassing = folium.FeatureGroup(name = "Trespassing")
Prostitution = folium.FeatureGroup(name = "Prostitution")
Homicide = folium.FeatureGroup(name = "Homicide")
Arson = folium.FeatureGroup(name = "Arson")
Robbery = folium.FeatureGroup(name = "Robbery")
Counterfeiting_Forgery = folium.FeatureGroup(name = "Counterfeiting/Forgery")
Sex_Offenses_Non_Forcible = folium.FeatureGroup(name = "Sex Offenses-Non Forcible")
Embezzlement = folium.FeatureGroup(name = "Embezzlement")
Liquor_Law_Violation = folium.FeatureGroup(name = "Liquor Law Violation")
Extortion_Blackmail = folium.FeatureGroup(name = "Extortion/Blackmail")
Family_Offenses_Non_Violent = folium.FeatureGroup(name = "Family Offenses Non Violent")

# create a dictionary with the 27 different crimes so they can be added to their unique layer
crime_wheel = {"Assault": Assault,
                      "Larceny/Theft": Larceny_Theft,
                "All Other Offenses":  All_Other_Offenses,
              "Sex Offenses-Forcible": Sex_Offenses_Forcible,
               "Fraud Offenses": Fraud_Offenses,
                          "Burglary": Burglary,
             "Weapon Law Violations": Weapon_Law_Violations,
                         "Vandalism": Vandalism,
            "Drug/Narcotic Offenses": Drug_Narcotic_Offenses,
       "Driving Under the Influence": Driving_Under_the_Influence,
               "Motor Vehicle Theft": Motor_Vehicle_Theft,
      "Pornography/Obscene Material": Pornography_Obscene_Material,
               "Disorderly Conduct": Disorderly_Conduct,
            "Kidnapping/Abduction": Kidnapping_Abduction,
        "Curfew/Loitering/Vagrancy": Curfew_Loitering_Vagrancy,
                  "Stolen Property": Stolen_Property,
                      "Trespassing": Trespassing,
                     "Prostitution": Prostitution,
                         "Homicide": Homicide,
                            "Arson": Arson,
                          "Robbery": Robbery,
           "Counterfeiting/Forgery": Counterfeiting_Forgery,
        "Sex Offenses-Non Forcible": Sex_Offenses_Non_Forcible,
            "Liquor Law Violation": Liquor_Law_Violation,
                    "Embezzlement": Embezzlement,
             "Extortion/Blackmail": Extortion_Blackmail,
"Family Offenses Non Violent": Family_Offenses_Non_Violent}

# create a colorwheel with 27 unique colors 
color_wheel = {"Assault": "rosybrown",
                      "Larceny/Theft": "gold",
                "All Other Offenses":  "orange",
              "Sex Offenses-Forcible": "teal",
               "Fraud Offenses": "saddlebrown",
                          "Burglary": "purple",
             "Weapon Law Violations": "violet",
                         "Vandalism": "slategray",
            "Drug/Narcotic Offenses": "seagreen",
       "Driving Under the Influence": "crimson",
               "Motor Vehicle Theft": "darkblue",
      "Pornography/Obscene Material": "turquoise",
               "Disorderly Conduct": "darkolivegreen",
            "Kidnapping/Abduction": "mintcream",
        "Curfew/Loitering/Vagrancy": "darkgoldenrod",
                  "Stolen Property": "coral",
                      "Trespassing": "thistle",
                     "Prostitution": "deeppink",
                         "Homicide": "darkgray",
                            "Arson": "dodgerblue",
                          "Robbery": "blue",
           "Counterfeiting/Forgery": "aquamarine",
        "Sex Offenses-Non Forcible": "darkkhaki",
            "Liquor Law Violation": "rebeccapurple",
                    "Embezzlement": "black",
             "Extortion/Blackmail": "lime",
"Family Offenses Non Violent": "rosybrown"}


# zoom in the map at an appropriate level
m = folium.Map(location = [38.5449, -121.7405], zoom_start = 15)
cols = ["crime_type", "y_coord", "x_coord", "occurrence_date"]
for name, latter, longer, clock in dat[cols].itertuples(index = False):
    popup = f"{name}, {clock}" # popup information
     # what the popup would look like
    crime = folium.Circle([latter,longer], color = color_wheel[name], popup= popup,fill = True, opacity = 1, fill_opacity = 1).add_to(crime_wheel[name])
Assault.add_to(m) # add all 27 different groups to the map
Larceny_Theft.add_to(m) 
All_Other_Offenses.add_to(m)
Fraud_Offenses.add_to(m) 
Sex_Offenses_Forcible.add_to(m) 
Burglary.add_to(m) 
Weapon_Law_Violations.add_to(m) 
Vandalism.add_to(m) 
Drug_Narcotic_Offenses.add_to(m) 
Driving_Under_the_Influence.add_to(m) 
Motor_Vehicle_Theft.add_to(m)
Pornography_Obscene_Material.add_to(m)
Disorderly_Conduct.add_to(m)
Kidnapping_Abduction.add_to(m)
Curfew_Loitering_Vagrancy.add_to(m)
Stolen_Property.add_to(m)
Trespassing.add_to(m)
Prostitution.add_to(m)
Homicide.add_to(m)
Arson.add_to(m)
Robbery.add_to(m)
Counterfeiting_Forgery.add_to(m)
Sex_Offenses_Non_Forcible.add_to(m)
Embezzlement.add_to(m)
Liquor_Law_Violation.add_to(m)
Extortion_Blackmail.add_to(m)
Family_Offenses_Non_Violent.add_to(m)
folium.LayerControl(collapsed = True).add_to(m)
fig = folium.Figure(width = 800, height = 500)
fig.add_child(m)

m.save("crime_map.html") # save the map
```
